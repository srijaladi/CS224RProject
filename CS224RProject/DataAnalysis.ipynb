```python
import matplotlib
from matplotlib import pyplot as plt
def plot():
    scores = np.loadtxt("score_history.txt")
    best_squares = np.loadtxt("best_square_history.txt")
    moves = np.loadtxt("move_history.txt")
    x_axis = np.arange(len(moves))
    
    plt.clf()
    plt.plot(x_axis, scores)
    plt.title("SCORE HISTORY OVER RUNS")
    plt.show()
    plt.clf()
    
    plt.clf()
    plt.plot(x_axis, best_squares)
    plt.title("BEST SQUARE REACHED HISTORY OVER RUNS")
    plt.show()
    plt.clf()
    
    plt.clf()
    plt.plot(x_axis, moves)
    plt.title("MOVES COMPLETED HISTORY OVER RUNS")
    plt.show()
    plt.clf()
    
plot()
```
